```python
def check(ch):
    if ch.isdigit():
        print("Entered char is a digit")
    elif ch.islower():
        print("Entered char is a lowercase letter")
    elif ch.isupper():
        print("Entered char is an uppercase letter")
    else:
        print("Entered char is a special character")

ch = input("Enter a character: ")
check(ch)
```

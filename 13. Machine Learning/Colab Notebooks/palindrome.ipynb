```python
# Function to check if a string is a palindrome
def is_palindrome(value):
    value = value.lower().replace(" ", "")  # Convert to lowercase & remove spaces
    return value == value[::-1]  # Compare with its reverse

# Taking user input
user_input = input("Enter a value: ")

# Checking if it's a palindrome
if is_palindrome(user_input):
    print("The entered value is a palindrome.")
else:
    print("The entered value is not a palindrome.")
```
